Regplot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example dataset
data = {
    'engine-size': np.random.randint(1000, 3000, 50),
    'price': np.random.randint(5000, 20000, 50),
    'Soil_Moisture': np.random.uniform(10, 50, 50),
    'Temperature_C': np.random.uniform(20, 40, 50)
}
df = pd.DataFrame(data)

# Regression plot
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0, )
plt.title('Engine Size vs Price')
plt.show()

# Correlation matrix
correlation = df[["Soil_Moisture", "Temperature_C"]].corr()
print(correlation)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset
data = {
    'Rainfall_mm': np.random.uniform(50, 300, 50),
    'Soil_Moisture': np.random.uniform(10, 50, 50)
}
df = pd.DataFrame(data)

# Regression plot
sns.regplot(x="Rainfall_mm", y="Soil_Moisture", data=df)
plt.ylim(0,)
plt.title('Rainfall vs Soil Moisture')
plt.show()

# Correlation matrix
correlation = df[["Rainfall_mm", "Soil_Moisture"]].corr()
print("Correlation Matrix:")
print(correlation)
```
